Year End Data Report

```python
import pandas as pd

# Load the data to see the initial few rows and understand its structure
file_path = '/content/Task 5.csv'
Year_End_Data = pd.read_csv(file_path)
Year_End_Data.head()
```

```python
#separate School A and B

School_A = Year_End_Data[Year_End_Data['school'] == 'School A']
School_B = Year_End_Data[Year_End_Data['school'] == 'School B']

School_A.head()
School_B.head()
```

```python
#Total Number of students

Sum_Students_SchoolA = School_A['student'].nunique()
Sum_Students_SchoolB = School_B['student'].nunique()

Sum_grade7_SchoolA = School_A[School_A['grade'] == 7]['student'].nunique()
Sum_grade8_SchoolA = School_A[School_A['grade'] == 8]['student'].nunique()

Sum_grade7_SchoolB = School_B[School_B['grade'] == 7]['student'].nunique()
Sum_grade8_SchoolB = School_B[School_B['grade'] == 8]['student'].nunique()



print(f"Number of Students in School A: {Sum_Students_SchoolA}")
print(f"Number of Students in Grade 7 in School A: {Sum_grade7_SchoolA}")
print(f"Number of Students in Grade 8 in School A: {Sum_grade8_SchoolA}\n\n")


print(f"Number of Students in School B: {Sum_Students_SchoolB}")
print(f"Number of Students in Grade 7 in School B: {Sum_grade7_SchoolB}")
print(f"Number of Students in Grade 8 in School B: {Sum_grade8_SchoolB}")
```

```python
#School A by Gender

School_A_male = School_A[School_A['gender'] == 'M']['student'].nunique()
School_A_female = School_A[School_A['gender'] == 'F']['student'].nunique()




print(f"Number of Male Students in School A: {School_A_male}")
print(f"Number of Female Students in School A: {School_A_female}")
```

```python
#view Math and Reading school performance percentage by school
# SCHOOL A
A_math_proficiency = School_A['is_prof_math'].mean()*100
A_reading_proficiency = School_A['is_prof_read'].mean()*100
A_math_growth = School_A['met_growth_math'].mean()*100
A_reading_growth = School_A['met_growth_read'].mean()*100

print(f"School A Math Proficiency: {A_math_proficiency:.2f}%"  )
print(f"School A Math Growth: {A_math_growth:.2f}% \n")

print(f"School A Reading Proficiency: {A_reading_proficiency:.2f}%")
print(f"School A Reading Growth: {A_reading_growth:.2f}% \n\n")

#SCHOOL B
B_math_proficiency = School_B['is_prof_math'].mean()*10
B_reading_proficiency = School_B['is_prof_read'].mean()*100
B_math_growth = School_B['met_growth_math'].mean()*100
B_reading_growth = School_B['met_growth_read'].mean()*100

print(f"School B Math Proficiency: {B_math_proficiency:.2f}%")
print(f"School B Math Growth: {B_math_growth:.2f}% \n ")

print(f"School B Reading Proficiency: {B_reading_proficiency:.2f}%")
print(f"School B Reading Growth: {B_reading_growth:.2f}%")
```

```python
#Visualization Math Proficiency

import matplotlib.pyplot as plt
import numpy as np

# Math Proficiency
plt.bar(['School A', 'School B'], [A_math_proficiency, B_math_proficiency], color=['skyblue', 'lightgreen'])

plt.xlabel('School')
plt.ylabel('Math Proficiency (%)')
plt.title('Math Proficiency By School')
plt.show()

# Math Growth
plt.bar(['School A', 'School B'], [A_math_growth, B_math_growth], color=['skyblue', 'lightgreen'])

plt.xlabel('School')
plt.ylabel('Math Proficiency (%)')
plt.title('Math Growth By School')
plt.show()
```

```python
#Reading Proficiency by School

#Visualization Math Proficiency

import matplotlib.pyplot as plt
import numpy as np

# Math Proficiency
plt.bar(['School A', 'School B'], [A_reading_proficiency, B_reading_proficiency], color=['skyblue', 'lightgreen'])

plt.xlabel('School')
plt.ylabel('Reading Proficiency (%)')
plt.title('Reading Proficiency By School')
plt.show()

# Math Growth
plt.bar(['School A', 'School B'], [A_reading_growth, B_reading_growth], color=['skyblue', 'lightgreen'])

plt.xlabel('School')
plt.ylabel('Reading Proficiency (%)')
plt.title('Reading Growth By School')
plt.show()
```

```python
# Reading and Math Met Growth by ELL
# Remove nulls and ensure proper filtering for 'ell' column

# Strip and lowercase
School_A['ell'] = School_A['ell'].str.lower().str.strip()
School_B['ell'] = School_B['ell'].str.lower().str.strip()

# Calculate growth rates, handling missing values with fillna(0)
ell_school_a = School_A[School_A['ell'] == 'yes']['met_growth_read'].fillna(0).mean()
ell_school_b = School_B[School_B['ell'] == 'yes']['met_growth_read'].fillna(0).mean()

print(f"School A Reading Growth for ELL Students: {ell_school_a:.2f}%")
print(f"School B Reading Growth for ELL Students: {ell_school_b:.2f}%")
```
